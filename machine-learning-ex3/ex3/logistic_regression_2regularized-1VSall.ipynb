```python
def sigmoid(z):
    return 1/(1+np.exp(-z))
```

```python
def loss(X, y, theta, llambda):
    return (-y.dot(np.log(sigmoid(X.dot(theta))))-(1-y).dot(np.log(1-sigmoid(X.dot(theta)))))/len(X) + llambda/2/len(X)*theta[1:].dot(theta[1:])
```

```python
def gradient_descent_step(X, y, theta, alfa, llambda):
    return theta - alfa / len(X) * (sigmoid(X.dot(theta)) - y).dot(X) - llambda/len(X)*theta
```

```python
# train

alfa = 3
llambda = 1
iterations = 1000
classifiers = 10

def train_kth_classifier(X, y, k):
    losses = np.zeros(iterations)
    X_train = np.c_[ np.ones(len(X)), X ] # add intercept
    y_train = np.where(y == k, 1, 0).flatten()
    theta = np.zeros(X_train.shape[1]) # init theta

    y_train.dot(np.log(sigmoid(X_train.dot(theta))))
    for i in range(iterations):
        theta = gradient_descent_step(X_train, y_train, theta, alfa, llambda)
        losses[i] = loss(X_train, y_train, theta, llambda)

    return theta, losses

all_losses = np.empty((0,iterations), int)
all_theta = np.empty((0,X.shape[1] + 1), int)

for i in range(classifiers):
    print("training classifier {}".format(i+1))
    theta, losses = train_kth_classifier(X, y, i+1)
    all_losses = np.append(all_losses, np.reshape(losses, (1, -1)), axis = 0)
    all_theta = np.append(all_theta, np.reshape(theta, (1, -1)), axis = 0)
```

```python
import matplotlib.pyplot as plt

for i in range(classifiers):
    plt.plot(all_losses[i])
plt.ylabel('loss')
plt.xlabel('iterations')
plt.show()
```

```python
# predict
X_predict = np.c_[ np.ones(len(X)), X ] # add intercept
len(np.amax(X_predict.dot(np.transpose(all_theta)), axis=1))
y_probabilities = sigmoid(X_predict.dot(np.transpose(all_theta)))
y_predicted = np.argmax(y_probabilities, axis=1) + 1
y_predicted
```

```python
def print_accuracy(y, y_predicted):
    TP = 0
    for i in range(len(y)):
        TP += np.where(y[i] == y_predicted[i],1,0)
    print("accuracy:   {}".format(np.round(TP/len(y),3)))
```

```python
print_accuracy(y, y_predicted)
```
