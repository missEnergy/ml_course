```python
def loss(X, y, theta):
    return (X.dot(theta) - y).dot(X.dot(theta) - y)/2/len(X)
```

```python
def gradient_descent_step(X, y, theta, alfa):
    n = len(X)
    temp0 = theta[0] - alfa / n * (X.dot(theta) - y).dot(X[: , 0])
    temp1 = theta[1] - alfa / n * (X.dot(theta) - y).dot(X[: , 1])
    theta[0] = temp0
    theta[1] = temp1
    return theta
```

```python
alfa = 0.01
iterations = 3000

theta = np.ones(2) # init theta
losses = np.zeros(iterations)

for i in range(len(losses)):
    gradient_descent_step(X, y, theta, alfa)
    losses[i] = loss(X, y, theta)
    
theta
```

```python
import matplotlib.pyplot as plt
plt.plot(losses)
plt.ylabel('loss')
plt.xlabel('iterations')
plt.show()
```
